### Python 메모리 모델의 핵심
- Python은 객체 참조 기반 언어다.
- 변수는 값을 담지 않는다.
- 객체를 "가르킨다"

```python
a = 10
b = a
```

```python
print(id(a))
print(id(b))
# 출력: 140734261727328 140734261727328
# a와 b는 같은 메모리 주소를 가리키고 있다.
```

### Immutable vs Mutable
| 구분        | 타입              | 특징    |
| --------- | --------------- | ----- |
| Immutable | int, str, tuple | 변경 불가 |
| Mutable   | list, dict, set | 변경 가능 |


### Immutable 예제

```python
a = 10
b = a
a = 20
```

```python
"""
처음:
[10] ← a, b

a = 20 실행 후:
[10] ← b
[20] ← a

값을 바꾼 게 아니라
새 객체를 생성한 것이다.
"""
```

```python
print(f"a: {a}, b: {b}")
# 출력: a: 20, b: 10
# 참조 값이 변경되지 않았다.
```

### Mutable 예제

```python
a = [1,2,3]
b = a 
b.append(4)
```

```python
print(f"a: {a}, b: {b}")
# 출력: a: [1, 2, 3, 4], b: [1, 2, 3, 4] a도 바뀐다. 같은 리스트 객체를 참조하기 떄문
```

### 실무에서 자주 나는 버그


```python
# 잘못된 코드
def add_item(lst):
    lst.append(100)

my_list = [1, 2, 3]
add_item(my_list)

print(my_list)
# 출력: [1, 2, 3, 100]
# 함수 안에서 수정했는데 밖에서도 바뀐다.
```

### 해결 방법 (복사)

```python
# 1. 얕은 복사
a = [1,2,3]
b = a.copy()
# b = a[:] 
```

```python
print(f"a: {a}, b: {b}")
# 출력: a: [1, 2, 3], b: [1, 2, 3]
b.append(4)
print(f"a: {a}, b: {b}")

# 중첩 객체는 복사안됨 (얕은 복사 함정)
a = [[1,2,3], [4,5,6]]
b = a.copy()
b[0].append(100)
print(f"a: {a}, b: {b}")
# 출력: a: [[1, 2, 3, 100], [4, 5, 6]], b: [[1, 2, 3, 100], [4, 5, 6]]
```

```python
# 2. 깊은 복사
import copy
b = copy.deepcopy(a)
# 완전 분리
# 단점은 성능비용
```

### 함수 기본값 함정 (실무에서 매우 중요)

```python
# 위험한 코드

def add_item(item, lst=[]):
    lst.append(item)
    return lst
print(add_item(1))
print(add_item(2))

#기본값 리스트는 한번만 생성된다.
```

```python
# 안전한 패턴
def add_item(item, lst=None):
    if lst is None:
        lst = []
    lst.append(item)
    return lst

print(add_item(1))
print(add_item(2))
```

### is vs ==
| 연산자 | 의미        |
| --- | --------- |
| ==  | 값 비교      |
| is  | 객체 동일성 비교 |


```python
a = [1,2]
b = [1,2]

print(a == b)  # True
print(a is b)  # False
```

### 메모리 설계 관점 정리
| 상황        | 선택       |
| --------- | -------- |
| 데이터 보호 필요 | deepcopy |
| 단순 리스트 처리 | copy     |
| 불변 데이터 설계 | tuple 사용 |
| 상태 공유 필요  | 참조 유지    |


### 데이터는 가능한 한 불변으로 설계

- 설정값 → tuple

- 상수 → 대문자

- 함수는 입력값을 직접 수정하지 말 것
